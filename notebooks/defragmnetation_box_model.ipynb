```julia
function defragmentation_event(V, fragmentation_mode)
    """
    V: volume of sphere
    k: size class. 0 if original
    fragmentation_mode: 1/2 if it breaks in two in a fragmentation event.
    """
    new_V = V/fragmentation_mode
#     k += 1
    
    return new_V
end 
```

```julia
function fragmentation(V, dt, fragmentation_mode, fragmentation_timescale)
    fragmentation_prob = exp(-dt/(fragmentation_timescale))
    rnd_num = rand()
    
    if rnd_num > fragmentation_prob
        V = defragmentation_event(V, k, fragmentation_mode)
    end
    
    return V
end
```

```julia
r = 1e-7 # 100nm
k = 0
sim_time = 365

V = 4/3*π*r^3
mode = 1/2
frag_timescale = 10
frag_mode = 1/2
k = 0
```

```julia
dt = 1/24
time = collect(0:dt:sim_time);
```

```julia
N = length(time)
```

```julia
radiuses = zeros(N)
volumes = zeros(N)

radiuses[1] = r
volumes[1] = V

for i in 1:N
    V = fragmentation(V, dt, frag_mode, frag_timescale)
    volumes[i] = V
    radiuses[i] = cbrt(3*V/(4*π))
end
```
