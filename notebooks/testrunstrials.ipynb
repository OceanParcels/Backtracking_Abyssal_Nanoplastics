```python
import numpy as np
import matplotlib.pyplot as plt
import gsw
import xarray as xr
import toolbox
import cartopy as cart
import cartopy.crs as ccrs
import cmocean.cm as cmo
import seaborn as sns
from glob import glob
import matplotlib
```

```python
G = xr.load_dataset('/storage/shared/oceanparcels/output_data/data_Claudio/tests/3BKNFP.nc')
```

```python
N = 999

fig,ax = plt.subplots()
ax.plot(G['time'][N], G['density'][N], c='blue', label='Radius')
ax.set_ylabel('density (m)', color='blue')
ax2 = ax.twinx()
ax2.plot(G['time'][N], G['v_s'][N], c='red', label='v_s')
ax2.set_ylabel('Sinking velocity (m/s)', color='red')

plt.title(r'$\lambda_f = 10$ days. No $K_z$')
```

```python
N = 12

fig,ax = plt.subplots()
ax.plot(G['time'][N], G['radius'][N], c='blue', label='Radius')
ax.set_ylabel('radius (m)', color='blue')
ax2 = ax.twinx()
ax2.plot(G['time'][N], G['v_s'][N], c='red', label='v_s')
ax2.set_ylabel('Sinking velocity (m/s)', color='red')

plt.title(r'$\lambda_f = 10$ days. No $K_z$')
```
